***Main Idea***: In 2018 over 100,000 developers took a survey on a popular website called Stack Overflow. Using seaborn visualizations, I wanted to spot some interesting trends among developers around the world such as comparing their educational background with employment status, discovering their career satisfaction rate, and see among which types of web developers, salaries are the highest. 

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

survey = pd.read_csv('survey_results.csv')
```

```python
plt.figure(figsize=(8,6))
colors = ["#8FA2FD", "#4374B3","#FF0B04","#339CFF","#1F45FC", "#3B5CFC"]
plot1 = sns.countplot(y="FormalEducation", hue="EmploymentStatus",data=survey, palette=colors)

plot1.set_title('Formal Education and Developer Employment Status')
plot1.set_ylabel('Formal Education')
plot1.set_xlabel('Number of People')
```

### Plot 1: Comparing Education level with Employment Status

This graph shows that the highest number of web developers and engineers who are currently working full-time, have received either a bachelors or a master degree. Although there are also people who never finished college, the number of those people employed full-time is lower compared to those with higher education degrees.

```python
sns.factorplot(y='YearsCodedJob', x='Salary',kind='bar', data=survey)
```

### Plot 2: Comparing Salaries of Different Web Developers

This graph shows that among the three categories of web developers around the world, more front-end web developers reported  lower salaries compared to the other 2 categories. based on this survey visualization, full stack web developers reported higher salaries based on the packed clustering of the dots from 25,000 to 125,000 U.S. dollars.

```python
plt.figure(figsize=(10,6))

plot2 = pd.melt(survey, "Gender", var_name="Web Developer Type")


# Draw a categorical scatterplot to show each observation
sns.swarmplot(x="WebDeveloperType", y="Salary",
              palette=["r", "c", "y"], data=survey).set_title('Comparing Salaries of Different Web Developers')
```

### Plot 3: Effects of Porgramming as a "Hobby" on Job & Career Satisfaction
   
These subplots show that in the range from 1 to 10, programmers who code as a hobby and contribute to open source projects, reported highest job and career satisfaction rates (10.0 points). However, the most common satisfaction rating for all groups, was around 7 to 7.5 points. 
    

```python
plot3= sns.pairplot(survey,hue="ProgramHobby", vars=["JobSatisfaction", "CareerSatisfaction"], height=4, aspect=.8,palette="RdBu_r")
plot3.fig.suptitle("Effects of Porgramming as a Hobby on Job & Career Satisfaction", y=1.08)
```
